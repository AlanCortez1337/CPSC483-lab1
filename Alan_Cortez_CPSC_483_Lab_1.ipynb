#### Part 1 data preparation

In the cell below, generate training data for your model.
<br> Generate a decent amount of training data in the interval of x.

<br> Since we are only approximating a function, data can be generated by plug in x value and get corresponding y value.
<br>You may use the way to generate data as in file Introduction_to_Artificial_Neural_Network.ipynb ([np.linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html))

```python

import matplotlib.pyplot as plt
import numpy as np

xc = np.linspace(-100, 100, 10001)

yc = xc * np.sin(xc**2 / 300)

plt.plot(xc, yc)
plt.show()
```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.

<br> You may use the model built in file Introduction_to_Artificial_Neural_Network.ipynb and change number of layers, number of neurons in each layer, activation function, loss function (loss), batch size, and/or epochs of the model. Or you can build your own model from scratch.

```python
# model 1
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten

X_train = xc
Y_train = yc

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(250,activation = "relu"))
model.add(Dense(250,activation = "LeakyReLU"))
model.add(Dense(250,activation = "ThresholdedReLU"))
model.add(Dense(250,activation = "PReLU"))
model.add(Dense(250,activation = "ELU"))
model.add(Dense(250,activation = "relu"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train, Y_train, batch_size = 128, epochs = 50, validation_split=0.6)

```

```python
# model 2
X_train_2 = xc
Y_train_2 = yc

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(200,activation = "relu"))
model.add(Dense(200,activation = "LeakyReLU"))
model.add(Dense(200,activation = "ThresholdedReLU"))
model.add(Dense(200,activation = "PReLU"))
model.add(Dense(200,activation = "ELU"))
model.add(Dense(200,activation = "relu"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train_2, Y_train_2, batch_size = 150, epochs = 60, validation_split=0.6)


```

```python
# model 3
X_train_3 = xc
Y_train_3 = yc

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(150,activation = "relu"))
model.add(Dense(150,activation = "LeakyReLU"))
model.add(Dense(150,activation = "ThresholdedReLU"))
model.add(Dense(150,activation = "PReLU"))
model.add(Dense(150,activation = "ELU"))
model.add(Dense(150,activation = "relu"))
model.add(Dense(150,activation = "LeakyReLU"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train_3, Y_train_3, batch_size = 200, epochs = 50, validation_split=0.6)

```

#### Part 3 model evaluation

In the cells below, eval your model with training data, test data ( with [model.evaluate()](https://keras.io/api/models/model_training_apis/#evaluate-method) ), and plot your predict result on the same plot with the plot of goal function.

```python
train_loss_1, train_mae_1 = model.evaluate(X_train, Y_train, 128)
train_loss_2, train_mae_2 = model.evaluate(X_train, Y_train, 150)
train_loss_3, train_mae_3 = model.evaluate(X_train, Y_train, 200)

print(f'MODEL #1\nTraining loss: {train_loss_1:.4f}, Training MAE: {train_mae_1:.4f}\n===========================================')
print(f'MODEL #1\nTraining loss: {train_loss_2:.4f}, Training MAE: {train_mae_2:.4f}\n===========================================')
print(f'MODEL #1\nTraining loss: {train_loss_3:.4f}, Training MAE: {train_mae_3:.4f}\n===========================================')

Y_predict = model.predict(xc)

plt.plot(xc, Y_predict, label="Predictions")
plt.plot(xc, yc, label="Original")
plt.show()
```
